```python
import cv2 
import matplotlib.pyplot as plt
```

### Read image

```python
img = cv2.imread('bears.jpg')
if img is None:
    print('image not found')
else:
    cv2.imshow('wooow so Cute!', img)
    cv2.waitKey()
    cv2.destroyAllWindows()
    plt.figure(figsize=[10,4])
    plt.subplot(121);plt.imshow(img);plt.title('BGR')
    plt.subplot(122);plt.imshow(img[...,::-1]);plt.title('RGB')
```

### convert to gray first way

```python
img = cv2.imread('bears.jpg')
if img is None:
    print('image not found')
else:
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    cv2.imshow('Gray Bears', gray)
    cv2.waitKey()
    cv2.destroyAllWindows()
    plt.imshow(gray, cmap ='gray')
```

### convert to gray second way (the Best way)

```python
gray = cv2.imread('bears.jpg',0)
if gray is None:
    print('image not found')
else:
    cv2.imshow('Gray Bears', gray)
    cv2.waitKey()
    cv2.destroyAllWindows()
    plt.imshow(gray, cmap ='gray')
```

### Save image

```python
cv2.imwrite('gray bears.png' , gray)
```
